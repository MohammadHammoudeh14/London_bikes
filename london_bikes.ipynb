```python
import pandas as pd
```

```python
# loading CSV data to a Pandas DataFrame
path = '/content/drive/MyDrive/Data /london_bikes.csv'
bikes = pd.read_csv(path)
bikes.head()
```

```python
bikes.shape
```

```python
# count the unique values in the weather_code column
bikes.weather_code.value_counts()
```

```python
# Count the unique values in season column
bikes.season.value_counts()
```

```python
# changing ciloumns names
new_names = { 'timestamp':'time',
             'cnt':'count',
              't1':'temp_real_c',
              't2':'temp_feels_like_C',
              'hum':'humidity_percent',
              'wind_speed':'wind_speed_kph',
              'weather_code':'weather',
              'is_holiday':'is_holiday',
              'is_weekend':'is_weekend',
              'season':'season'
              }

# Renaming the columns to the specified column names
bikes.rename(new_names, axis=1, inplace=True)
```

```python
# changing the humidity values to percentage (i.e. a value between 0 and 1)
bikes.humidity_percent = bikes.humidity_percent / 100
```

```python
# Crear a season dictionary to change numeric values to STR
season_dict = { '0.0':'spring',
               '1.0':'summer',
                '2.0':'autumn',
                '3.0':'winter'}

# Crear a weather dictionary to change numeric values to STR
weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'
}

# cahnge season datatype to string
bikes.season = bikes.season.astype("str")
# mapping the values 0-3 to the actual written seasons
bikes.season = bikes.season.map(season_dict)

# cahnge weather datatype to string
bikes.weather = bikes.weather.astype("str")
# mapping the values 0-3 to the actual written seasons
bikes.weather = bikes.weather.map(weather_dict)
```
